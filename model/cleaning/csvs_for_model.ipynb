```python
# Import
import pandas as pd
```

# PULLING OUT AVERAGES FOR DIFFERENT TYPES OF PROPERTIES 

```python
# Reading in cleaned csv as a DataFrame
df = pd.read_csv("output_data/clean_house_crime_data.csv")
df.head()
```

```python
# Grouping by type of house, applying mean aggregate
grouped_type_df = df.groupby("Type").mean()
grouped_type_df
```

```python
# Export to csv into 'data folder'
grouped_type_df.to_csv("../../data/type_averages.csv")
```

# PULLING OUT COLUMNS BASED ON SUBURB 

Distance from CBD, crime rate and average increase in growth of each suburb

```python
suburb_df = pd.read_csv("output_data/data_including_suburb.csv")
suburb_df.head()
```

```python
# Dropping duplicates, so only one of each suburb
suburb_df = suburb_df.drop_duplicates('Suburb')
suburb_df.head()
```

```python
print(f" The length of the DatafFrame is {len(suburb_df)} rows. Number of unique suburbs is {suburb_df['Suburb'].nunique()}.")
```

```python
# Pulling out columns of interest
sub_crime_dist_df = suburb_df[["Suburb", "Distance", "Crime"]]
sub_crime_dist_df
```

```python
# Reading in csv as a DataFrame
increase_df = pd.read_csv("sourced_data/Average_Suburb_Price_Increase.csv")
increase_df.head()
```

```python
# Pulling out suburb column and average yearly increase % for each
increase_df = increase_df.loc[:, ('Median Unit Price - 2019','Unnamed: 15')]
increase_df.head()
```

```python
# Renaming columns
increase_df.columns = ('Suburb', 'Average_increase')
```

```python
# Changing suburb column to lower case
increase_df['Suburb'] = increase_df['Suburb'].str.lower()
increase_df.head()
```

```python
# Merging two dataframes on suburb
combined_df = pd.merge(sub_crime_dist_df, increase_df, how = 'inner', on='Suburb')
combined_df.head()
```

```python
# Resetting index as suburb names
sub_crime_dist_increase_df = combined_df.set_index('Suburb')
sub_crime_dist_increase_df.head()
```

```python
# Export to csv
sub_crime_dist_increase_df.to_csv("../../data/suburb_crime_distance_avincrease.csv")
```
